```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['text.usetex'] = True
```

```python
def fixed_method_iteration(x, y, x0, n):
    xk = [x0]
    x_est = x0
    for _ in range(n):
        x_est = np.interp(x_est, x, y)
        xk.append(x_est)
    return xk
```

```python
def iteration_plot(pts):
    h = 0
    its_x = []
    its_y = []
    i = 0
    while i < len(pts)-2:
        its_x.append(pts[i])
        its_y.append(pts[i])
        its_x.append(pts[i])
        its_y.append(pts[i+1])
        its_x.append(pts[i+1])
        its_y.append(pts[i+1])
        its_x.append(pts[i+1])
        its_y.append(pts[i+2])
        i += 2
    return np.array(its_x), np.array(its_y)
```

```python
def plot_fix_method_iteration(ax, x, y, n, x0, func_text):
    # Move the left and bottom spines to x = 0 and y = 0, respectively.
    #ax.spines[["left", "bottom"]].set_position(("data", 0))
    # Hide the top and right spines.
    #ax.spines[["top", "right"]].set_visible(False)

    ax.plot(x,y, 'r', zorder=5)
    ax.plot(x,x, 'k--', zorder=1)
    ax.text(1, 4, func_text, color="r", fontsize=16)

    fix_point_its = fixed_method_iteration(x, y, x0, n)
    its_x, its_y = iteration_plot(fix_point_its)

    for i in range(n):
        ax.arrow(its_x[i], its_y[i], its_x[i+1]-its_x[i], its_y[i+1]-its_y[i], head_width=0.05, head_length=0.1, fc='k', ec='k', length_includes_head=True, zorder=10)

    ax.set_xlim(-0.3,5)
    ax.set_ylim(-0.3,5)
```

```python
# 100 linearly spaced numbers
x = np.linspace(+0.0001,5,100)
y = 1 + 2/x
n = 10

fig1, ax1 = plt.subplots()
plot_fix_method_iteration(ax1, x, y, n, 4, r"$1+\frac{2}{x}$")

fig2, ax2 = plt.subplots()
y = x**2
plot_fix_method_iteration(ax2, x, y, n, 1.3, r"$x^2$")
# save
fig1.show()
fig2.show()
```
