### MÃ©todo del punto fijo

```python
def MetodoPuntoFijo(f,x0,n):
    x = x0
    for k in [1..n]:
        x = f(*x)
    return x
```

#### Ejemplos

##### Ejemplo 1: coseno

```python
f(x)=(cos(x),)
```

```python
MetodoPuntoFijo(f,(0.5,),20)[0]
```

```python
def MetodoPuntoFijoIteraciones(f,x0,n):
    x = [x0] + [0]*n
    for k in [1..n]:
        x[k] = f(*x[k-1])
    return x
```

```python
def MetodoPuntoFijoIteracionesB(f,x0,n):
    x = [x0] + [0]*n
    for k in [1..n]:
        x[k] = f(*x[k-1])
    for k in [0..n]:
        x[k] = x[k][0]
    return x
```

```python
[(x, f(x)[0]) for x in MetodoPuntoFijoIteracionesB(f,(0.5,),10)]
xf = [(x, f(x)[0]) for x in MetodoPuntoFijoIteracionesB(f,(0.5,),10)][-1]
xf
```

```python
d1 = line([(x, f(x)[0]) for x in MetodoPuntoFijoIteracionesB(f,(4,),10)],color='green',linestyle='--')
plot(f(x)[0],(-1,2)) + d1 + point(xf,color='red',size=30)
```
